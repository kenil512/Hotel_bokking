#### Finding null values in both no_of_adults and no_of_children columns and drop those values from data frame

```python
df = df[~((df['no_of_adults'].isnull()) & (df['no_of_children'].isnull()))]
df = df[~((df['no_of_adults']== 0) & (df['no_of_children'].isnull()))]
df = df[~((df['no_of_adults'].isnull()) & (df['no_of_children']== 0))]
df.no_of_adults = df.no_of_adults.fillna(0)
df.no_of_children = df.no_of_children.fillna(0)
df['total_members'] = df.no_of_adults + df.no_of_children
```

#### Number of total memberes paying average price per room by various market segment

```python
ax = sns.scatterplot(data=df, x='total_members', y='avg_price_per_room', hue="market_segment_type", style= 'market_segment_type', alpha = 0.8)
plt.title('Number of total memberes paying average price per room by various market segment')
plt.show()
```

#### this scatterplot represent that lead_time vs average price per room by booking status

```python
ax = sns.scatterplot(data=df, x = "lead_time", y = 'avg_price_per_room', hue = 'booking_status')
plt.title('lead time Vs. average price per room by various booking status')
plt.show()
```

#### Average booking status by lead_time

```python
sort_booking_status = df.groupby('booking_status')['lead_time'].mean()
sort_booking_status = pd.DataFrame(sort_booking_status)
ax = sns.barplot(data=sort_booking_status, x = sort_booking_status.index , y = sort_booking_status.lead_time.to_list(), palette ="dark:#5A9_r")
plt.xlabel('Booking_status')
plt.ylabel('Lead_time (Days)')
plt.title('Average lead time by booking status')
plt.show()
# this bar graph tell us more the lead time higher the cancelation, less the lead time lower the cancelation
# average lead time for non cnaeled booking status only 50% than canceled booking status
# if customers has higher lead time, this meaning that highest probebility to cancelation. Due to this reason hotel booking company may be charged high for this type of client
```

#### Average price per room by various room type reserved and type of meal plan

```python
sorted_room_type = df[~df.room_type_reserved.isna()]
sorted_newdf = sorted_room_type[~sorted_room_type.avg_price_per_room.isna()]
sorted_newdf = sorted_newdf[(sorted_newdf.avg_price_per_room > 40) & (sorted_newdf.avg_price_per_room < 330)]
ax = sns.boxplot(data = sorted_newdf, x = 'room_type_reserved' , y ="avg_price_per_room", palette ="ch:2", hue = 'type_of_meal_plan')
plt.title('Average price per Room by various Room type reserved')
plt.xticks(rotation=45)
plt.show()
# it is important to remove all data which having null values
# when we think about average price per room it should be more than 40 
# this boxplot tell us median average price per room with type of meal plan
# meal plan 2 which has higher average price per room than others except room type 5
# most of the people preffered meal plan 1 & meal plan 2 
# room type 1 & room type 2 has low average price per room than rest of all room types
```
